# Counting cars on static images

The code uses YOLO_V3 library from cvlib which recognizes some common objects, including cars, used for this specific exercise in less than 10 lines of code.

Note that depending on image quality and car crowded images, results may have a low accuracy.

### Import required libraries

```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
#Read image from disk
im = cv2.imread('./data/im1.jpg')
#Detect objects using 'detect_common_objects' module from cvlib.
bbox, label, conf = cv.detect_common_objects(im)
#Draw a box over the recognized objects
output_image = draw_bbox(im, bbox, label, conf)
#Show image (with bounding box included)
plt.imshow(output_image)
plt.show()
#Print the number of cars (filtered by label 'car').
print('Number of cars in the image is '+ str(label.count('car')))
```
